```python
import matplotlib.pyplot as pp
from mpl_toolkits.mplot3d import Axes3D as axes
import numpy as np

YRATIO = 39.47
XRATIO = 1881600
yRange = np.arange(YRATIO * .9, YRATIO * 1.1, .1)
xRange = np.arange(XRATIO * .9, XRATIO * 1.1, 10)

x = np.random.choice(xRange, 100, replace=True)
y = np.random.choice(yRange, 100, replace=True)
z = x * y

fig = pp.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_title("EBIT & E")
ax.set_zlabel("EBIT * EBIT factors")
ax.set_xlabel("EBIT")
ax.set_ylabel("EBIT factors")
```
