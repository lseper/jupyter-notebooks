```python
import matplotlib.pyplot as pp
import numpy as np
import math
import scipy.integrate as sci
```

## Question 1:

```python
def exp_decay(l, x):
    return l * math.exp(-1 * l * x)
# generate a range of values 0 - 10, increasing by .1
x1 = np.arange(0, 10, .1)
# copy the array so we don't modify the original
x1Copy = np.copy(x1)
x2Copy = np.copy(x1)
# compute the outputs for the given input values
for i in range(x1.size):
    x1Copy[i] = exp_decay(1, x1Copy[i])
    x2Copy[i] = exp_decay(.5, x2Copy[i])
    
# create the plot
pp.plot(x1Copy, label='位 = 1.0')
pp.plot(x2Copy, label='位 = 0.5')
pp.legend(loc = 'best')
pp.xlabel('x')
pp.ylabel(r'f(x) = $\lambda - e^{-\lambda*x}$')
pp.xticks(np.arange(0, 100, 10), [i / 100 for i in range(100) if i % 10 == 0]) 
pp.title('Exponential Decay')
pp.show()
```

#### Is the function symmetric?

No, the function is not symmetric.

```python
def cum_exp_decay(l, x):
    return 1 - math.exp(-l * x)

# create range of values
xVals = np.arange(0, 50, .1)
xVals2 = np.arange(0, 50, .1)

# get outputs from inputs using list comprehensions
yVals = [cum_exp_decay(1, val) for val in xVals]
yVals2 = [cum_exp_decay(0.5, val) for val in xVals2]

# create the plots
pp.plot(xVals, yVals, label='位 = 1.0')
pp.plot(xVals2, yVals2, label='位 = 0.5')
pp.legend(loc = 'best')
pp.xlabel('x')
pp.ylabel(r'f(x) = 1 - $e^{-\lambda*x}$')
pp.title('Cumulative Exponential Decay')
pp.ylim(0, 1.5)
pp.show()
```

## Question 2:

```python
# create inputs
xVals = np.arange(-8, 8, .1)
xVals2 = np.arange(-8, 8, .1)

# create output values
yVals = [exp_decay(1, val * val) for val in xVals]
yVals2 = [exp_decay(0.5, val * val) for val in xVals2]

# create the plot
pp.plot(xVals, yVals, label='位 = 1.0')
pp.plot(xVals2, yVals2, label='位 = 0.5')
pp.legend(loc = 'best')
pp.xlabel('x')
pp.ylabel(r'f(x) = 1 - $e^{-\lambda*x^2}$')
pp.title('Squared Exponential Decay')
pp.ylim(0, 1.2)
pp.show()
```

## Question 4:

```python
def logistic(x, r):
    if(x <= 0):
        return 0.5
    else:
        return r * logistic(x-1, r) * (1 - logistic(x-1, r))
# inputs
xVals = np.arange(0, 20, 1)

# outputs
y1Vals = [logistic(val, 1.5) for val in xVals]
y2Vals = [logistic(val, 3.58) for val in xVals]

# plot
pp.plot(xVals, y1Vals, label=r'r = 1.5')
pp.plot(xVals, y2Vals, label='r = 3.58')
pp.legend(loc = 'best')
pp.xlabel('x')
pp.ylabel(r'$x_{i+1}=r*x_{i}*(1-x_{i})$')
pp.title('Discrete Logistics Equation')
pp.ylim(0, 1.3)
pp.show()
```

#### What is going on here?

The function is highly sensitive to initial changes, so it is characterized as chaotic. It seems that as the value for r increases, the range that the function oscilates between is greater.
