## Question 6:

```r
sq_exp <- function(x, o, u){
    temp <- (x - u) / o
    temp <- -1/2 * (temp^2)
    return(exp(temp))
}

# create inputs
x <- seq(-10, 10, 0.1)
# set up sigma and mu values
sigmaValues = c(1, 1, 2)
muValues = c(0, 1, 0)

# plot the standard function
sigma <- sigmaValues[1]
mu <- muValues[1]
plot(x, sq_exp(x, sigma, mu), col="red", type="l", main="squared exponential function")

# plot the shifted function
mu <- muValues[2]
sigma <- sigmaValues[2]
lines(x, sq_exp(x, sigma, mu), col="blue", type="l")

# plot the spread function
mu <- muValues[3]
sigma <- sigmaValues[3]
lines(x, sq_exp(x, sigma, mu), col="green", type="l")
legend("topright", c("standard", "shifted", "spread"), fill = c("red", "blue", "green"))
```

```r
norm_sq_exp = function(x, o, u) {
    temp = 1 / (sqrt(2 * pi) * o)
    return(sq_exp(x, o, u) * temp)
}

# plot the standard
sigma <- sigmaValues[1]
mu <- muValues[1]
plot(x, norm_sq_exp(x, sigma, mu), col="red", type="l", main="standardized squared exponential function")

# plot the shifted
mu <- muValues[2]
sigma <- sigmaValues[2]
lines(x, norm_sq_exp(x, sigma, mu), col="blue", type="l")

# plot the spread
mu <- muValues[3]
sigma <- sigmaValues[3]
lines(x, norm_sq_exp(x, sigma, mu), col="green", type="l")
legend("topright", c("standard", "shifted", "spread"), fill = c("red", "blue", "green"))
```

## Question 10:

```r
# set up some helper functions for our recursive newton's method
norm_cubic = function(x){
    return(x^3 - (2 * x) - 5)
}

deriv_cubic = function(x) {
    return (3*x^2 - 2)
}

# set up newton's method function
recursive_call = function(x){
    return (x - (norm_cubic(x)/deriv_cubic(x)))
}

# iterate through it for 20 values
val = recursive_call(0)
for(i in 0:20) {
    val = recursive_call(val)
}
# print the approximation yielded
s = sprintf("Newton's Method Approximation: %.7f", val)
print(s)
```
