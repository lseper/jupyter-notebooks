## Question 4

```python
n = 1000
p = 0.2
q = 1 - p
mean = n * p
stdDev = (n * p * q) ** (1/2)

# approximate binomial dist as a normal distribution
binomNorm = stats.norm(mean, stdDev).pdf

print(f'Probability that between 170 and 185 telephones will be white: {round(scint.quad(binomNorm, 170, 185)[0], REG_ROUND)}')
print(f'Probability that between 210 and 225 telephones will be white: {round(scint.quad(binomNorm, 210, 225)[0], REG_ROUND)}')
```

## Question 10

```python
mu = 5
sigma = [0.5, 1, 2]

def normal(m, o, minVal, maxVal):
    normalPDF = stats.norm(m, o).pdf
    retVals = []
    for i in range(minVal, maxVal):
        retVals.append(scint.quad(normalPDF, i, i + 1)[0])
    return retVals

print(normal(mu, sigma[0], 0, 10))

xVals = np.arange(0, 10, 1)
yVals = normal(mu, sigma[1], 0, 10)
pp.bar(xVals, yVals)
pp.ylim(0, 1)
pp.title('Normal Distribution w/ μ = 5, σ = 1')
pp.xlabel('X')
pp.ylabel('P(X = x)')
pp.show()

print(f'μ = {mu}')
print(f'σ = {sigma[1]}')

xVals = np.arange(0, 10, 0.1)
pp.plot(xVals, stats.norm.pdf(xVals, mu, sigma[0]), label = 'μ = 5, σ = 0.5')
pp.plot(xVals, stats.norm.pdf(xVals, mu, sigma[1]), label = 'μ = 5, σ = 1')
pp.plot(xVals, stats.norm.pdf(xVals, mu, sigma[2]), label = 'μ = 5, σ = 2')
pp.ylim(0, 1)
pp.title('Normal Distributions with various σ values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As $\sigma$ increases, the probabilities appears to flatten and become wider, stretching probabilities more evenly. Conversely, as $\sigma$ decreases, the probabilities appears to narrow, condensing probabilities towards the $\mu$ of the normal distribution

```python
xVals = np.arange(0, 10, 0.1)
pp.plot(xVals, stats.norm.cdf(xVals, mu, sigma[0]), label = 'μ = 5, σ = 0.5')
pp.plot(xVals, stats.norm.cdf(xVals, mu, sigma[1]), label = 'μ = 5, σ = 1')
pp.plot(xVals, stats.norm.cdf(xVals, mu, sigma[2]), label = 'μ = 5, σ = 2')
pp.ylim(0, 1)
pp.title('Cumulative Normal Distributions with various σ values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As $\sigma$ increases, the cumulative probability rises and levels out smoothly. Conversely, as $\sigma$ decreases, the cumulative probability rises and levels off more sharply than if it weren't

## Question 11

```python
alpha = 12
beta = [0.25, 0.5, 1]

def gamma(a, b, minVal, maxVal):
    gammaPDF = stats.gamma(a=a, scale=b).pdf
    retVals = []
    for i in range(minVal, maxVal):
        retVals.append(scint.quad(gammaPDF, i, i + 1)[0])
    return retVals

print(gamma(alpha, beta[0], 0, 10))

xVals = np.arange(0, 10, 1)
yVals = gamma(alpha, beta[1], 0, 10)
pp.bar(xVals, yVals)
pp.ylim(0, 1)
pp.title('Beta Distribution w/ alpha = 12, beta = 0.5')
pp.xlabel('X')
pp.ylabel('P(X = x)')
pp.show()

print(f'μ = {alpha * beta[1]}')
print(f'σ = {alpha * beta[1] ** 2}')

xVals = np.arange(0, 25, 0.2)
pp.plot(xVals, stats.gamma.pdf(xVals, a=alpha, scale=beta[0]), label = 'alpha = 12, beta = 0.25')
pp.plot(xVals, stats.gamma.pdf(xVals, a=alpha, scale=beta[1]), label = 'alpha = 12, beta = 0.5')
pp.plot(xVals, stats.gamma.pdf(xVals, a=alpha, scale=beta[2]), label = 'alpha = 12, beta = 1')
pp.ylim(0, 1)
pp.title('Beta Distributions with various σ values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As the beta value increases, the probability values seems to flatten out, increasing much more gradually. Conversely, as the beta value decreases, the probability values increases and decreases more steeply reaching a higher peak.

```python
xVals = np.arange(0, 25, 0.2)
pp.plot(xVals, stats.gamma.cdf(xVals, a=alpha, scale=beta[0]), label = 'alpha = 12, beta = 0.25')
pp.plot(xVals, stats.gamma.cdf(xVals, a=alpha, scale=beta[1]), label = 'alpha = 12, beta = 0.5')
pp.plot(xVals, stats.gamma.cdf(xVals, a=alpha, scale=beta[2]), label = 'alpha = 12, beta = 1')
pp.ylim(0, 1)
pp.title('Cumulative Beta Distributions with various σ values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As beta increases, the cumulative probability increases more smoothly, and levels off more smoothly. Conversely, as beta decreases, the cumulative probability increases and decreases much more sharply, rather than gradually.

## Question 12

```python
a = 0
b = [1, 5, 10]

def uniform(a, b, minVal, maxVal):
    uniformPDF = stats.uniform(a, b).pdf
    retVals = []
    for i in range(minVal, maxVal):
        retVals.append(scint.quad(uniformPDF, i, i + 1)[0])
    return retVals

print(uniform(a, b[0], 0, 10))

xVals = np.arange(0, 10, 1)
yVals = uniform(a, b[1], 0, 10)
pp.bar(xVals, yVals)
pp.ylim(0, 1)
pp.title('Uniform Distribution w/ a = 0, beta = 5')
pp.xlabel('X')
pp.ylabel('P(X = x)')
pp.show()

print(f'μ = {1/2 * (a + b[1])}')
print(f'σ = {round(1/12 * ((b[1] - a) ** 2), SPEC_ROUND)}')

xVals = np.arange(0, 10, 0.2)
pp.plot(xVals, stats.uniform.pdf(xVals, a, b[0]), label = 'a = 0, b = 1')
pp.plot(xVals, stats.uniform.pdf(xVals, a, b[1]), label = 'a = 0, b = 5')
pp.plot(xVals, stats.uniform.pdf(xVals, a, b[2]), label = 'a = 0, b = 10')
pp.ylim(0, 1)
pp.title('Uniform Distributions with various b values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As b increases, the probability of each individual x decreases, but is spread out over a larger range. As b decreases, the probability of each individual x increases, and is condensed to a smaller range.

```python
xVals = np.arange(0, 10, 0.2)
pp.plot(xVals, stats.uniform.cdf(xVals, a, b[0]), label = 'a = 0, b = 1')
pp.plot(xVals, stats.uniform.cdf(xVals, a, b[1]), label = 'a = 0, b = 5')
pp.plot(xVals, stats.uniform.cdf(xVals, a, b[2]), label = 'a = 0, b = 10')
pp.ylim(0, 1)
pp.title('Cumulative Uniform Distributions with various b values')
pp.legend()
pp.ylabel("P(x)")
pp.xlabel("x")
pp.show()
```

As b increases, the slope is more gradual before leveling off at 1. As b decreases, the slope is much steeper before leveling off at 1.
